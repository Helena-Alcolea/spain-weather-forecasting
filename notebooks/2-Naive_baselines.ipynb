# Naive Baselines — Weather Forecasting for Spanish SYNOP Stations

This notebook establishes naive baselines for temperature and precipitation forecasting over four prototype SYNOP stations. These baselines define the minimum performance threshold that any trained model must exceed to demonstrate genuine predictive value.

**Forecast targets:** daily maximum temperature (tmax), minimum temperature (tmin), and precipitation (prec)  
**Forecast horizons:** t+1, t+3, t+7 days  
**Prototype stations:** Córdoba Aeropuerto, Burgos, Puerto de Navacerrada, Palma Puerto  
**Validation scheme:** Expanding Window CV — 20 annual windows, minimum 15 years of training data, identical to the ML baseline scheme

## 1. Baseline Methods

Two naive baselines are evaluated:

**Persistence:** the forecast for day t+h is the observed value at day t — the last known observation before the forecast horizon. This is the simplest possible baseline and captures the short-term autocorrelation of meteorological variables.

**Climatology (seasonal mean):** the forecast for any given day is the historical mean for that calendar day, computed exclusively from the training data of each fold. This captures the annual cycle without using any recent observations.

Neither baseline involves any model training. Their purpose is to establish a lower performance bound: a model that does not beat both baselines provides no useful information beyond what is already available from historical patterns.

## 5. Key Findings

### 5.1 Temperature: a clear crossover between baselines

A consistent pattern emerges across all four stations and both temperature variables:

- **At t+1**, persistence outperforms climatology in every station. Temperature has strong short-term autocorrelation: tomorrow will closely resemble today, and persistence captures that signal directly.
- **At t+3 and t+7**, climatology outperforms persistence in every station. As the forecast horizon increases, the memory of current conditions fades and the seasonal cycle dominates. Climatology encodes that seasonal signal by design.

This crossover defines the **predictability horizon of naive methods**: somewhere between t+1 and t+3, short-term memory becomes less useful than long-term seasonal patterns.

The implication for model evaluation is direct: any trained model must beat persistence at t+1 and climatology at t+3 and t+7. The harder bar varies by horizon.

### 5.2 LightGBM clears both baselines at all horizons

The ML baseline (LightGBM) outperforms both naive baselines for temperature across all stations and all horizons. The margin over climatology narrows at t+7 — particularly for Burgos (LightGBM 3.58°C vs climatology 3.64°C) — but LightGBM consistently remains below both naive thresholds. This validates that the ML models extract genuine predictive signal beyond naive pattern-matching.

### 5.3 Precipitation: naive baselines fail entirely

Both baselines produce negative R² for precipitation across all stations and all horizons. A negative R² means the baseline performs worse than simply predicting the sample mean — a result that reflects the physical nature of daily precipitation rather than a methodological failure.

Precipitation is governed by mesoscale and synoptic atmospheric dynamics (frontal systems, convective triggering, orographic lifting) that are not captured by local station observations alone. Neither the memory of yesterday's rain nor the seasonal average provides a reliable signal for tomorrow's precipitation at daily resolution.

This finding motivates two directions in subsequent phases:
1. **Deep learning:** LSTM and GRU architectures may capture sequential dependencies in station data that naive methods cannot.
2. **Problem reformulation:** if DL does not substantially improve precipitation forecasting, the task will be reformulated as binary classification (rain / no rain) with a physically motivated threshold.
